```python
min = -20
max = 70
K = 30
barriere = 20
zoom = 5
dt = 1000

st = np.linspace(min,max,dt)
st_zoom = np.linspace(min,K-barriere,dt)
st_pointilles = np.linspace(K-barriere,K,dt)
st_digit = np.linspace(K-barriere,max,dt)
st_barriere =  np.linspace(-barriere+zoom,zoom,dt)

plt.figure(figsize=(12,6))

plt.plot(st,st - K , linewidth=1.5,color = "black", label = "sous-jacent")
plt.plot(st_zoom, st_zoom-K+zoom, linewidth=2.3, color = "dodgerblue")
plt.plot(st_digit, np.zeros(len(st_barriere))+zoom, linewidth=2.3, color = "dodgerblue")
plt.plot(st_pointilles, st_pointilles-K+zoom, "--", linewidth=1.4, color = "dodgerblue")
plt.plot(K-barriere+np.zeros(len(st_barriere)), st_barriere, "--",linewidth=1.5,  color='dimgray', alpha=0.5, label="Barrière")

#pour une meilleure visibilité
plt.plot(np.zeros(len(st))+K,st-K, '--', color='dimgray', alpha=0.3, linewidth=1.2)
plt.plot(st,np.zeros(len(st)), linewidth=0.5, color='dodgerblue')
plt.plot(np.zeros(len(st)),st-K, linewidth=0.5, color='steelblue')

#plt.axis([min, max, -20, 20])
plt.xticks([K], ["Strike"])
plt.grid(axis="y", linewidth = 0.2)
plt.title('Autocall')
plt.legend()
plt.show()
```
