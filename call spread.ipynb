```python
st = np.arange(4,10)
k1 = 6; k2=7
call1 = np.asarray([0 if k1>x else x-k1 for x in st]) #on achète
call2 =  np.asarray([0 if k2>x else k2-x for x in st]) #on vend

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(st,call1,"--", label="Call 1", alpha = 0.6)
ax.plot(st,call2,"--", label="Call 2", alpha = 0.6)
ax.plot(st,call1+call2, label="Call spread", color="purple", linewidth = 3)

plt.grid(linewidth=0.6)
plt.xticks([k1,k2], ['$k_1$','$k_2$'], fontsize = 11)

ax.set_ylim(-3,4)
ax.legend(prop={'size': 12})
ax.set_ylabel("Prix du sous-jacent",size=12)
ax.set_xlabel("Sous jacent",size=12)
ax.set_title("Call spread", size=14)
plt.savefig("figOut/Call spread")
```

```python
#version put
st = np.arange(15)
k1 = 6; k2=7
put1 = np.asarray([0 if k1<x else x-k1 for x in st]) #on vend
put2 =  np.asarray([0 if k2<x else k2-x for x in st])+0.1 #on achète

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(st,put1,"--", label="Call 1", linewidth=1)
ax.plot(st,put2,"--", label="Call 2", linewidth=1)
ax.plot(st,put1+put2, label="Call spread", color="purple")

plt.grid(linewidth=0.2)
plt.xticks([k1,k2], ['$k_1$','$k_2$'])

ax.legend()
ax.set_ylabel("Prix")
ax.set_xlabel("S")
ax.set_title("Call spread")
```

```python
st = np.linspace(0,14,29)
k = 7
call = np.asarray([0 if k>x else x-k for x in st]) #on achète
put =  np.asarray([0 if k<x else k-x for x in st]) #on achète

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(st,call,"x", label="Call")
ax.plot(st,put,".", label="Put")
ax.plot(st,call+put, label="sum Call Put",color="purple")

plt.grid(linewidth=0.2)
plt.xticks([k], ['$k$'])

ax.legend()
ax.set_ylabel("Prix")
ax.set_xlabel("S")
ax.set_title("*************")
```

```python
st = np.arange(15)
k1 = 7; k2=10; k3=15
call = np.asarray([0 if k1>x else x-k1 for x in st]) #on achète
put1 =  np.asarray([0 if k2<x else k2-x for x in st]) #on achète
put2 =  np.asarray([0 if k3<x else x-k3 for x in st]) #on vend
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(st,call,"--", label="Call 1", linewidth=1)
ax.plot(st,put1,"--", label="Put 1", linewidth=1)
ax.plot(st,put2,"--", label="Put 2", linewidth=1)

ax.plot(st,call+put1+put2, label="Call spread", color="purple")

plt.grid(linewidth=0.2)
plt.xticks([k1,k2,k3], ['$k_1$','$k_2$','$k_3$'])

ax.legend()
ax.set_ylabel("Prix")
ax.set_xlabel("S")
ax.set_title("Call spread")
```

```python
st = np.linspace(0,12,2000)
k1 = 5; k2=6

digit = np.asarray([0 if k2>x else 1 for x in st])#on achète

call1 = np.asarray([0 if k1>x else x-k1 for x in st]) #on achète
call2 =  np.asarray([0 if k2>x else k2-x for x in st]) #on vend


fig, ax = plt.subplots(figsize=(10,6))

ax.plot(st,digit, label="Digit", alpha=0.8)
ax.plot(st,call1+call2+0.01, "--",label="Digit lissé par un call spread", color="purple")

plt.grid(linewidth=0.6)
plt.xticks([k1,k2], ['$k_1$', '$k_2$'], fontsize=11)

ax.legend(prop={'size': 12})
ax.set_ylabel("Pourcentage",size=12)
ax.set_xlabel("Sous-jacent",size=12)
ax.set_title("Digit\n",size=14)
plt.savefig("figOut/Digit")
```
