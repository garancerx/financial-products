```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mean = 410
k = 300
spot = np.arange(1000)
sj = norm.pdf(spot,mean,150)
sj2 = norm.pdf(spot,mean,50)
call = np.asarray([0 if k>x else (x-k) for x in spot]) #on ach√®te
minimum = np.minimum(sj,call/50000)
minimum2 = np.minimum(sj2,call/50000)
```

```python
plt.subplots(figsize=(12,6))

plt.plot(spot, call/50000, color="purple", label = "Call")

plt.plot(spot, sj, color='turquoise', label = "2 ans", linewidth=1)
plt.plot(spot, sj2, color='gold', label = "2 semaines", linewidth=1)

plt.fill_between(spot, minimum, color='#539ecd', alpha = 0.2, label = "Esp√©rance 2 ans")
plt.fill_between(spot, minimum2, color='khaki', alpha = 0.2, label = "Esp√©rance 2 semaines")

plt.grid(linewidth=0.3)
plt.xticks([mean,k], ['$S_t$','strike'], fontsize=10)

plt.ylim(-0.0002,0.0085)
plt.xlabel('Spot')
plt.ylabel('Probabilit√©\n')
plt.legend()

plt.savefig("figOut/normal_distribution.png")
plt.show()
```

```python
mean = 310
k = 250
spot = np.linspace(-50,700,1000)
sj = norm.pdf(spot,mean,120)
sj2 = norm.pdf(spot,mean,40)
call = np.asarray([0 if k>x else 3*(x-k) for x in spot]) #on ach√®te
minimum = np.minimum(sj,call/65000)
minimum2 = np.minimum(sj2,call/65000)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(spot, sj, color='turquoise', label = "2 ans", linewidth=1)
ax.plot(spot, sj2, color='gold', label = "2 semaines", linewidth=1)

ax.fill_between(spot, minimum, color='#539ecd', alpha = 0.2, label = "Esp√©rance 2 ans")
ax.fill_between(spot, minimum2, color='khaki', alpha = 0.3, label = "Esp√©rance 2 semaines")

ax2 = ax.twinx()

ax2.plot(spot, call+32, color="purple", label = "Call")

ax.grid(linewidth=0.3)
ax.set_xticks([mean,k], ['$S_t$','strike'], fontsize=11)

#ax.set_ylim(-0.0002,0.0085)
ax.set_xlabel('Spot\n', size= 12)
ax.set_ylabel('Probabilit√©\n', size=12)
ax2.set_ylabel('\n Payoff', size=12)
ax2.set_ylim(0,700)
plt.figlegend(loc = 'center left', ncol=1,bbox_to_anchor = (0.15,0.75))

plt.savefig("figOut/normal_distribution.png")
```
